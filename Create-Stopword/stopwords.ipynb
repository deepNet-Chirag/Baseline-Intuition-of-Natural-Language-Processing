```python
from nltk.probability import FreqDist
from nltk.tokenize import word_tokenize
import os
import numpy as np
import pandas as pd
```

```python
with open("hindi essays.txt","r",encoding = "utf-8") as f:
    content = f.read()
```

```python
print(content)
```

```python
content_list = word_tokenize(content)
```

```python
content_list[0:10]
```

```python
content_dist = FreqDist(content_list)
```

```python
content_dist
```

```python
print("Total unique words",len(content_dist))
```

```python
content_dist.most_common(30)
```

```python
stopwords = []
```

```python
for tup in content_dist.most_common(29):
    stopwords.append(tup[0])
```

```python
stopwords.remove('भारत')
stopwords.remove('समाज')
stopwords.remove('साहित्य')
```

```python
print(stopwords)
print("Length of stopwords",len(stopwords))
```

```python
with open("hindi stories.txt",'r',encoding='utf-8') as f:
    text = f.read()
```

```python
print(text)
```

```python
string = ""

for word in word_tokenize(text):
    if word not in stopwords:
        string += word + " "
```
