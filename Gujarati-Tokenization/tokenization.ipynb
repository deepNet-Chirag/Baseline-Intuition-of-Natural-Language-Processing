```python
import numpy as np
import pandas as pd
import nltk
from nltk.tokenize import sent_tokenize , word_tokenize
```

```python
with open("story_guj.txt","r", encoding = "utf-8") as f:
    content = f.read()
```

```python
print(content)
```

 # Using NLTK 

```python
#Word Tokenization

words_token = word_tokenize(content)
print(words_token)

print("\nTotal number of unique tokens",len(words_token))

print("\nTotal number of unique tokens",len(set(words_token)))
```

```python
#Sentence Tokenization

sents_token = sent_tokenize(content)
print(sents_token)
```

# Using Split function

```python
#Word tokenization using split function

wt = content.split(" ")
print(wt)

print("\nTotal number of unique tokens",len(wt))

print("\nTotal number of unique tokens",len(set(wt)))
```

```python
# Sentence tokenization using split function
st = content.split(". ")
print(st)
```
