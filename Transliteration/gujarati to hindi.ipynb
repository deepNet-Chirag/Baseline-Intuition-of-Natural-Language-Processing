```python
with open("story_guj.txt","r",encoding = "utf-8-sig") as f:
    content = f.read()
```

```python
print(content)
```

```python
#Difference between gujarati and hindi character in unicode is 384
#So here we subtract 384 from unicode value of gujarati character

string = ""

for c in content:
    try:
        y = ord(c) - 384
        y = chr(y)
        string += y
    except:
        string += " "
```

```python
with open("story_hin.txt","w",encoding="utf-8") as f:
    f.write(string)
```
